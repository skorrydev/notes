## Shodan - İnternete Bağlı Cihazlar İçin Arama Motoru

Shodan, kullanıcıların internete bağlı çeşitli cihaz türlerini (web kameraları, yönlendiriciler, sunucular vb.) çeşitli filtreler kullanarak aramasına olanacak tanıyan bir arama motorudur. Genellikle siber güvenlik araştırmaları ve savunmasız cihazların tespiti için kullanılır.

İşte bazı örnek Shodan sorguları ve ilgili komutlar:

*   **RTSP Akış Tespiti ve İzleme:**
    *   `ffplay -rtsp_transport tcp rtsp:/ip`: Bu komut, TCP üzerinden bir RTSP (Real-Time Streaming Protocol) akışını oynatmak için `ffplay` aracını kullanır. `/ip` kısmı, Shodan üzerinden bulunan gerçek bir IP adresi ve port ile değiştirilmelidir.
    *   `ffplay -rtsp_transport tcp rtsp:/117.212.151.181:554`: Belirli bir IP ve porttan (554 RTSP için varsayılandır) akış oynatmak için somut bir örnek.
    *   **RTSP için Shodan Sorgusu:** Bu tür akışları bulmak için Shodan'da `port:554 rtsp` gibi bir sorgu kullanabilirsiniz.

*   **VNC Sunucu Keşfi (Uzaktan Masaüstü Kontrolü):**
    *   `VNC screenshot.label:ics country:"US" default port:5900 TCP`: Bu sorgu, Amerika Birleşik Devletleri'nde bulunan, ekran görüntüsü etiketi "ICS" (Endüstriyel Kontrol Sistemleri) olan ve varsayılan 5900 portunu kullanan VNC sunucularını arar.
    *   `-"kaitlyn.codes" AND -"resolv.moe" vnc screenshot.label:loggedin screenshot.label:desktop`: Giriş yapılmış ve masaüstü görünen VNC sunucularını ararken, belirli adresleri (bal küpü/honeypot olma ihtimaline karşı) filtreler.

*   **Android Debug Bridge (ADB) Keşfi:**
    *   `"Android debug bridge (ADB)" -Authentication port:5555`: Port 5555 üzerinde çalışan ve kimlik doğrulama gerektirmeyen (potansiyel zafiyet) ADB servislerini arar.

## Sosyal Mühendislik Teknikleri

Bireyleri gizli bilgileri ifşa etmeye veya kendi çıkarlarına olmayan eylemleri gerçekleştirmeye ikna etmek için kullanılan manipülasyon yöntemleridir.

*   **Eavesdropping (Gizli Dinleme):** Konuşmacıların bilgisi dışında gizlice dinleme.
*   **Shoulder Surfing (Omuz Atlatma):** Birinin şifre veya PIN gibi hassas bilgilerini girerken omzunun üzerinden izleme.
*   **Dumpster Diving (Çöp Karıştırma):** Kötü amaçlarla kullanılabilecek değerli bilgileri bulmak için çöpleri karıştırma.
*   **Impersonation (Taklit Etme):** Yetkisiz erişim veya bilgi elde etmek için başka biriymiş gibi (BT desteği, yönetici vb.) davranma.

# Siber Güvenlik Keşif ve Analiz Rehberi

Bu yazı, bir hedef sistem üzerinde bilgi toplama (reconnaissance) ve analiz süreçlerinde kullanılan temel teknikleri ve araçları detaylandırmaktadır.

---

## 1. Brute Force (Kaba Kuvvet) Saldırıları
Brute Force, bir parolanın veya gizli bir verinin, olası tüm kombinasyonların sistematik olarak denenmesiyle ele geçirilmesi yöntemidir.
*   **Sözlük Saldırıları (Dictionary Attacks):** En yaygın kullanılan şifrelerin bulunduğu listeler üzerinden deneme yapılır.
*   **Hibrit Saldırılar:** Sözlükteki kelimelere rakam veya sembol ekleyerek yapılan denemelerdir.

## 2. DNS Keşif Araçları
Alan adı sistemleri (DNS), hedef ağın mimarisini anlamak için kritik bilgiler sunar.
*   **dnsenum:** Bir domain hakkında MX, NS ve A kayıtlarını sorgular, alt alan adlarını kaba kuvvetle arar ve zone transfer (AXFR) denemesi yapar.
*   **dnsrecon:** DNS kayıtlarını toplamak için kullanılan çok fonksiyonlu bir araçtır. Wildcard tespiti ve SRV kayıtları sorgulama gibi gelişmiş özelliklere sahiptir.

## 3. Alt Alan Adı Keşfi (Subdomain Finder)
Bir kurumun ana sitesi kadar güvenli olmayan alt alan adlarını (test, dev, mail vb.) bulmak saldırı yüzeyini genişletir.
*   **Subfinder:** Pasif kaynakları kullanarak çok hızlı bir şekilde alt alan adlarını toplayan modüler bir araçtır.
*   **Findomain:** Hız odaklıdır ve sertifika şeffaflığı (Certificate Transparency) kayıtlarını kullanarak güncel domainleri yakalar.
*   **Sublist3r:** Birçok farklı arama motorundan ve kaynaktan veri çekerek sonuçları birleştiren klasik bir araçtır.

## 4. Ağ Haritalama Yöntemleri
Ağın fiziksel ve mantıksal yapısını belirlemek için kullanılır.
*   **Tracerouting:** Bir paketin hedefe giderken geçtiği router'ları ve izlediği yolu gösterir. Ağın topolojisini anlamaya yardımcı olur.
*   **Ping Sweep:** Bir IP aralığındaki hangi cihazların aktif olduğunu (ayakta olduğunu) ICMP paketleri göndererek tespit etme yöntemidir.
*   **Port Tarama (Port Scanning):** Hedef cihazdaki açık portları ve bu portlarda çalışan servisleri (HTTP, SSH, FTP) belirler. En popüler aracı **Nmap**'tir.
*   **Banner Grabbing:** Açık portlardaki servislerin versiyon bilgilerini (örneğin Apache 2.4.41) çekerek bilinen açıklara karşı analiz yapmayı sağlar.
*   **OS Fingerprinting:** Hedef sistemin TCP/IP yığınındaki karakteristik tepkilere bakarak hangi işletim sistemini (Linux, Windows, Cisco iOS) kullandığını tahmin eder.
